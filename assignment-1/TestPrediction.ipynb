```python
from __future__ import division

import pickle

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
from torch.autograd import Variable
# from torch.utils.data import TensorDataset, DataLoader

from convnet3_deep2 import ConvNet

%matplotlib inline
```

```python
test_data = pickle.load(open("data/test.p", "rb"))
test_data = test_data.test_data.numpy()
test_data = np.array([(x.flatten()/255.0).reshape(1,28,28) for x in test_data])
test_data = torch.from_numpy(test_data).float().resize_(len(test_data),1,28,28)
```

```python
valid_data = pickle.load(open('data/generated_valid_data_norm.p', 'rb'))
valid_data = torch.from_numpy(valid_data).float().resize_(len(valid_data),1,28,28)

valid_label = pickle.load(open('data/generated_valid_labels.p', 'rb'))
valid_label = torch.from_numpy(valid_label).long()
```

```python
model = ConvNet(output_dim=10, dropout=0.3)
```

```python
model = torch.load('saved_models/best_cnn_d25b16e89acc9891.model')
```

```python
model.eval()
```

```python
def predict(m, x_val):
    x = Variable(x_val, requires_grad=False)
    output = m.forward(x)
    return output.data.numpy().argmax(axis=1)
```

```python
pred_y = predict(model, valid_data)
```

```python
100. * np.mean(pred_y == valid_label.numpy())
```

```python
pred_test = predict(model, test_data)
```

```python
test_df = pd.DataFrame(pred_test, columns=['label'])
test_df.reset_index(inplace=True)
test_df.rename(columns={'index': 'ID'}, inplace=True)
```
